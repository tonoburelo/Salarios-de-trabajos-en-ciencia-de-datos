```python
#extraer el archivo del archivo zip descargado
zipfile_name = 'data-science-job-salaries.zip'
with zipfile.ZipFile(zipfile_name, 'r') as file:
    file.extractall()
```

```python
#leer el archivo csv como un dataframe de pandas
salarios_ds = pd.read_csv('ds_salaries.csv')
```

```python
#dejar solo las columnas que voy a utilizar de la tabla
salarios_ds = salarios_ds[['work_year', 'experience_level', 'employment_type', 'job_title', 'salary', 'salary_currency', 'salary_in_usd', 'employee_residence', 'remote_ratio', 'company_location', 'company_size']]
```

```python
#Contar cuantos valores son únicos en la columna job_title
len(salarios_ds['job_title'].unique())
```

```python
#contar cuantos valores tienen cada valor unico de la columna job_title
salarios_ds['job_title'].value_counts()
```

```python
#convertir los datos de la columna salary_in_usd en pesos mexicanos multiplicando por 17
salarios_ds['salario_en_pesos_mex'] = salarios_ds['salary_in_usd'].multiply(17)
```

```python
#especificar los nombre de columnas que quiero utilizar
nuevas_columnas = {
    'work_year':'año_trabajo',
    'experience_level': 'nivel_experiencia',
    'employment_type':'tipo_empleo',
    'job_title':'nombre_del_trabajo',
    'salary': 'salario',
    'salary_currency':'moneda_salario',
    'salary_in_usd':'salario_en_dolares',
    'employee_residence':'residencia_del_empleado',
    'remote_ratio':'proporción_remota',
    'company_location':'locación_compañia',
    'company_size':'tamaño_compañia'
}

#renombrar las columnas a las columnas especificadas
salarios_ds.rename(nuevas_columnas, axis=1, inplace=True)
```
